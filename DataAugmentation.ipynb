```python
from numpy import expand_dims
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
import os
from os import walk
from PIL import Image
from numpy import shape
import cv2
import numpy as np
from matplotlib import cm
```

```python
def shiftImage(img, counter, path, sampleNo, graphs = False):

    # convert to numpy array
    data = img_to_array(img)
    
    # expand dimension to one sample
    samples = expand_dims(data, 0)
    
    # create image data augmentation generator
    datagen = ImageDataGenerator(width_shift_range=0.2, height_shift_range=0.2)
    
    # prepare iterator
    it = datagen.flow(samples, batch_size=1)
    
    # generate samples and plot
    for i in range(sampleNo):
        
        if graphs == True:
            # define subplot
            pyplot.subplot(330 + 1 + i)
        
        # generate batch of images
        batch = it.next()

        # convert to unsigned integers for viewing
        image = batch[0].astype('uint8')
        
        if graphs == True:
            # plot raw pixel data
            pyplot.imshow(image,cmap='gray')
        
        image = np.reshape(image,(32,32))
        
        image = Image.fromarray(image, 'L')
        image.save(path+"\\"+str(counter)+".jpg")
        counter += 1
    pyplot.show()
    return counter
```

```python
def rotateImage(img, counter, path, sampleNo, graphs = False):

    # convert to numpy array
    data = img_to_array(img)
    
    # expand dimension to one sample
    samples = expand_dims(data, 0)
    
    # create image data augmentation generator
    datagen = ImageDataGenerator(rotation_range=45)
    
    # prepare iterator
    it = datagen.flow(samples, batch_size=1)
    
    # generate samples and plot
    for i in range(9):
        
        if graphs == True:
            # define subplot
            pyplot.subplot(330 + 1 + i)
        
        # generate batch of images
        batch = it.next()

        # convert to unsigned integers for viewing
        image = batch[0].astype('uint8')
        
        if graphs == True:
            # plot raw pixel data
            pyplot.imshow(image,cmap='gray')
            
        image = np.reshape(image,(32,32))
        
        image = Image.fromarray(image, 'L')
        image.save(path+"\\"+str(counter)+".jpg")
        counter += 1
    pyplot.show()
    return counter
```

```python
def main():
    os.mkdir("Characters\\JPEGAugmented\\")
    os.mkdir("Characters\\JPEGAugmented\\Shifted\\")
    os.mkdir("Characters\\JPEGAugmented\\FinalData\\")
    
    subfolders = [ f.path for f in os.scandir("Characters\JPEGResized") if f.is_dir() ]

    for subfolder in subfolders:
        
        _, _, filenames = next(walk(subfolder))
        counter = 0
        splitName = str(subfolder).split("\\")
        os.mkdir("Characters\\JPEGAugmented\\Shifted\\"+splitName[2])
        
        for filename in filenames:
            image = cv2.imread(subfolder+"\\"+filename, cv2.IMREAD_UNCHANGED)
            path  = "Characters\\JPEGAugmented\\Shifted\\"+splitName[2]
            counter = shiftImage(image,counter,path, 10)
            
    print("Vertical and Horizontal Shifts Done!")
    
    subfolders = [ f.path for f in os.scandir("Characters\JPEGAugmented\Shifted") if f.is_dir() ]
    
    for subfolder in subfolders:
        
        _, _, filenames = next(walk(subfolder))
        counter = 0
        splitName = str(subfolder).split("\\")
        os.mkdir("Characters\\JPEGAugmented\\FinalData\\"+splitName[3])
        
        for filename in filenames:
            image = cv2.imread(subfolder+"\\"+filename, cv2.IMREAD_UNCHANGED)
            #print(image.shape)
            path  = "Characters\\JPEGAugmented\\FinalData\\"+splitName[3]
            counter = rotateImage(image,counter,path, 10)
            
    print("Rotations Done!")
    print("Data Augmentation Complete!!!")
```

```python
main()
```
